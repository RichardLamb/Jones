```python
%matplotlib inline
import pickle
import json
import pymongo

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
pickle_in = open("businesses.pickle","rb")
businesses = pickle.load(pickle_in)
print(len(businesses))
#print(json.dumps(businesses[0], indent=2))
```

```python
plt.figure(figsize=(20,10))
plt.title('Businesses')
m = Basemap(projection='lcc', resolution='c',
            width=10E6, height=10E6, 
            lat_0=37, lon_0=-121,)
m.drawcoastlines()
m.drawstates()
m.fillcontinents(color='coral',lake_color='aqua')
m.drawmapboundary(fill_color='aqua')

for vendor in businesses:
    if vendor['coordinates']['latitude'] is not None:
        x, y = m(vendor['coordinates']['longitude'], vendor['coordinates']['latitude'])
        m.plot(x, y, 'bo', markersize=2)
```
